```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#reading data
df = pd.read_csv('/Users/Spencer/Desktop/data_projects/project1/python/movies.csv')
```

```python
#open dataset
df.head()
```

```python
# see if there are any nulls
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing)*100))
```

```python
# checking data types 
df.dtypes
```

```python
# replace empty cells with 0 for budget and gross column
df.budget.fillna(0, inplace=True)
df.gross.fillna(0, inplace=True)
# changing data type float to int for budget and gross column
df["budget"] = df["budget"].astype("int64")
df["gross"] = df["gross"].astype("int64")
```

```python
df
```

```python
# year and release column don't match up
df.released.replace(r'\s+',np.nan,regex=True).replace('',np.nan)
df['yearcorrect'] = df['released'].astype(str).str[0:4]
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#scatter plot budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show
```

```python
df.head()
```

```python
# plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# looking at correlation
# method default is pearson others are kendall and spearman
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# company correlation
df_num = df

for col_name in df_num.columns:
    if (df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes

df_num.head()
```

```python
correlation_matrix = df_num.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
